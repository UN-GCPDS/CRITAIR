````python
"""
DEPENDENCIES CONFIGURATION - BERTSCORE EVALUATION
================================================

Import of all necessary libraries for comparative analysis
of AI models using BertScore as the main metric.
"""

# ===================================================================================================
# EVALUATION AND METRICS LIBRARIES
# ===================================================================================================

from bert_score import BERTScorer  # 🎯 Main semantic evaluation metric

# ===================================================================================================
# DATA ANALYSIS LIBRARIES
# ===================================================================================================

import pandas as pd               # 📊 Structured data manipulation and analysis
import statistics                 # 📈 Statistical calculations (mean, deviation, etc.)
from datetime import datetime     # ⏰ Date and timestamp handling

# ===================================================================================================
# VISUALIZATION LIBRARIES
# ===================================================================================================

import matplotlib.pyplot as plt   # 📈 Charts and visualizations creation

# ===================================================================================================
# FILE LIBRARIES
# ===================================================================================================

from openpyxl import load_workbook  # 📄 Excel files reading with experimental results

print("✅ All dependencies imported successfully")
print("🎯 System ready for BertScore evaluation")
````

````python
"""
BERTSCORE EVALUATION FUNCTION
============================

Implementation of the main function to calculate semantic similarity
between target responses and AI model predictions.
"""

def get_bert_score(targets, predictions, scorer):
    """
    Calculates semantic similarity between target sentences and predictions using BertScore.
    
    BertScore is a metric that evaluates generated text quality by comparing
    contextual token representations using pre-trained language models.
    
    Args:
        targets (List[str]): List of target sentences (ground truth)
        predictions (List[str]): List of sentences predicted by the model
        scorer (BERTScorer): Configured BertScore evaluator instance
    
    Returns:
        float: Average BertScore F1 score (0.0 - 1.0)
               - Values close to 1.0 indicate high semantic similarity
               - Values close to 0.0 indicate low semantic similarity
    
    Notes:
        - Uses F1 score that combines precision and recall
        - The mean() function calculates the average of all evaluated pairs
        - item() extracts the numerical value from PyTorch tensor
    """
    try:
        # Calculate precision, recall and F1 for each target-prediction pair
        precision, recall, f1_scores = scorer.score(targets, predictions)
        
        # Return the mean of F1 score as main metric
        bert_score_mean = f1_scores.mean().item()
        
        return bert_score_mean
        
    except Exception as e:
        print(f"❌ Error calculating BertScore: {e}")
        return 0.0  # Return 0 in case of error

print("✅ get_bert_score function configured correctly")
print("📊 Metric: Average F1 Score for semantic evaluation")
````

````python
"""
CONFIGURACIÓN DEL EVALUADOR BERTSCORE
====================================

Inicialización del scorer BertScore con configuración optimizada para español.
"""

print("🔧 Configuring BertScore evaluator...")

# Configure BertScore for Spanish evaluation with baseline rescaling
scorer = BERTScorer(
    lang="es",                    # 🇪🇸 Language: Spanish
    rescale_with_baseline=True    # 📊 Baseline normalization for better comparisons
)

print("✅ BERTScorer configured successfully")
print("📋 Configuration:")
print("   • Language: Spanish (es)")
print("   • Baseline rescaling: Enabled")
print("   • Metric: Contextualized F1 Score")
print("   • Underlying model: Based on pre-trained transformers")
````

````python
"""
EXPERIMENTAL DATA LOADING - UNSTRUCTURED QUESTIONS
=================================================

Loading dataset with experimental results for unstructured questions.
"""

print("📊 LOADING EXPERIMENTAL DATA - UNSTRUCTURED QUESTIONS")
print("=" * 70)

try:
    # ⚠️ PRIVACY NOTE: This path references experimental data
    # In a production environment, this data would be in a secure location
    file_path = "/content/drive/MyDrive/Thesis/Experimental SetUp/bitacora_chatbot_recomendación_wo_think.xlsx"
    
    print("📂 Loading experimental results file...")
    print(f"   File: {file_path.split('/')[-1]}")
    print(f"   Sheet: No structure questions")
    
    # Load unstructured questions data
    df_results = pd.read_excel(
        file_path,
        sheet_name="No structure questions"
    )
    
    print(f"✅ Data loaded successfully")
    print(f"📋 Dataset dimensions: {df_results.shape}")
    print(f"   • Rows (questions): {df_results.shape[0]}")
    print(f"   • Columns (models + metadata): {df_results.shape[1]}")
    
    # Show dataset information
    if not df_results.empty:
        print(f"\n📊 DATASET INFORMATION:")
        print(f"   • Questions evaluated: {len(df_results)}")
        print(f"   • Available columns: {', '.join(df_results.columns[:5])}...")
        
        # Verify target responses column
        if "Respuesta Target" in df_results.columns:
            print(f"   • Target responses available: ✅")
        else:
            print(f"   • Target responses available: ❌")
    
    # Show dataset preview
    df_results
    
except FileNotFoundError:
    print("❌ Error: Data file not found")
    print("   Verify that the Excel file is available at the specified path")
    df_results = pd.DataFrame()  # Empty dataset to avoid errors
    
except Exception as e:
    print(f"❌ Error loading data: {e}")
    df_results = pd.DataFrame()
````

````python
"""
BERTSCORE EVALUATION - UNSTRUCTURED QUESTIONS
============================================

Comparative analysis of all AI models for unstructured questions
using BertScore as semantic similarity metric.
"""

print("🔬 STARTING BERTSCORE EVALUATION - UNSTRUCTURED DATA")
print("=" * 70)

# Initialize lists to store results
bertscore_by_model = []
times_by_model = []

# Extract target responses (ground truth)
if not df_results.empty and "Respuesta Target" in df_results.columns:
    targets = list(df_results["Respuesta Target"])
    print(f"✅ Target responses loaded: {len(targets)}")
    
    print(f"\n🔄 PROCESSING {len(models)} MODELS...")
    print("─" * 50)
    
    # Evaluate each model
    for i, model in enumerate(models):
        print(f"🤖 Evaluating: {model}")
        
        try:
            # Extract model predictions (column i+2 because first ones are metadata)
            predictions = list(df_results.iloc[:, i+2])
            
            # Calculate BertScore
            print(f"   📊 Calculating BertScore...")
            bert_score = get_bert_score(targets, predictions, scorer)
            bertscore_by_model.append(bert_score)
            
            # Extract inference times (column i+12)
            print(f"   ⏱️  Calculating average times...")
            times = list(df_results.iloc[:, i+12])
            
            # Clean and process times (remove invalid values)
            valid_times = [
                float(x) for x in times
                if x != '' and x is not None and not isinstance(x, datetime)
            ]
            
            if valid_times:
                time_mean = statistics.mean(valid_times)
                times_by_model.append(time_mean)
            else:
                times_by_model.append(float('inf'))  # Infinite time if no valid data
                
            # Show model results
            print(f"   ✅ BertScore: {bert_score:.4f}")
            print(f"   ⚡ Average time: {time_mean:.3f}s")
            print(f"   📊 Valid measurements: {len(valid_times)}/{len(times)}")
            
        except Exception as e:
            print(f"   ❌ Error processing {model}: {str(e)[:50]}...")
            bertscore_by_model.append(0.0)
            times_by_model.append(float('inf'))
        
        print("─" * 50)
    
    print(f"🎉 EVALUATION COMPLETED")
    print(f"   • Models evaluated: {len(bertscore_by_model)}")
    print(f"   • BertScores calculated: {sum(1 for score in bertscore_by_model if score > 0)}")
    print(f"   • Valid times: {sum(1 for time in times_by_model if time != float('inf'))}")
    
else:
    print("❌ Cannot evaluate models: insufficient data")
    bertscore_by_model = [0.0] * len(models)
    times_by_model = [float('inf')] * len(models)
````

````python
"""
RESULTS CONSOLIDATION - UNSTRUCTURED DATA
=========================================

Creation of DataFrame with consolidated BertScore and time results.
"""

print("📊 CONSOLIDATING RESULTS - UNSTRUCTURED QUESTIONS")
print("=" * 60)

# Crear DataFrame con resultados
bert_time_df = pd.DataFrame({
    "Model": models,
    "BertScore": bertscore_by_model,
    "Inference Time": times_by_model
})

print("✅ DataFrame de resultados creado")
print(f"📋 Estructura: {bert_time_df.shape[0]} modelos × {bert_time_df.shape[1]} métricas")

# Mostrar estadísticas rápidas
if not bert_time_df.empty:
    valid_scores = bert_time_df[bert_time_df['BertScore'] > 0]
    valid_times = bert_time_df[bert_time_df['Inference Time'] != float('inf')]
    
    print(f"\n📈 ESTADÍSTICAS GENERALES:")
    print(f"   • Modelos con BertScore válido: {len(valid_scores)}/{len(bert_time_df)}")
    print(f"   • Modelos con tiempo válido: {len(valid_times)}/{len(bert_time_df)}")
    
    if len(valid_scores) > 0:
        print(f"   • BertScore promedio: {valid_scores['BertScore'].mean():.4f}")
        print(f"   • Mejor BertScore: {valid_scores['BertScore'].max():.4f}")
        print(f"   • Peor BertScore: {valid_scores['BertScore'].min():.4f}")
    
    if len(valid_times) > 0:
        print(f"   • Tiempo promedio: {valid_times['Inference Time'].mean():.3f}s")
        print(f"   • Modelo más rápido: {valid_times['Inference Time'].min():.3f}s")
        print(f"   • Modelo más lento: {valid_times['Inference Time'].max():.3f}s")

# Mostrar el DataFrame completo
bert_time_df
````

````python
"""
CÁLCULO DE BALANCE CALIDAD/VELOCIDAD
===================================

Cálculo de la métrica de balance que combina BertScore y tiempo de inferencia.
"""

print("⚖️  CALCULANDO BALANCE CALIDAD/VELOCIDAD")
print("=" * 50)

# Calcular balance: Tiempo/BertScore (menor es mejor)
# - Valores bajos indican buen balance (alta calidad, tiempo razonable)
# - Valores altos indican mal balance (baja calidad o tiempo excesivo)
try:
    bert_time_df["Balance"] = bert_time_df["Inference Time"] / bert_time_df["BertScore"]
    
    # Manejar casos especiales
    bert_time_df["Balance"] = bert_time_df["Balance"].replace([float('inf'), -float('inf')], 999999)
    
    print("✅ Métrica de balance calculada")
    print("\n📊 INTERPRETACIÓN DEL BALANCE:")
    print("   • Valores bajos (< 10):  Excelente balance calidad/velocidad")
    print("   • Valores medios (10-50): Balance aceptable") 
    print("   • Valores altos (> 50):   Balance deficiente")
    
    # Mostrar estadísticas del balance
    valid_balance = bert_time_df[bert_time_df["Balance"] < 999999]
    
    if not valid_balance.empty:
        print(f"\n📈 ESTADÍSTICAS DE BALANCE:")
        print(f"   • Modelos con balance válido: {len(valid_balance)}")
        print(f"   • Mejor balance: {valid_balance['Balance'].min():.2f}")
        print(f"   • Peor balance: {valid_balance['Balance'].max():.2f}")
        print(f"   • Balance promedio: {valid_balance['Balance'].mean():.2f}")
        
        # Identificar mejor modelo por balance
        best_balance_idx = valid_balance['Balance'].idxmin()
        best_model = bert_time_df.loc[best_balance_idx, 'Model']
        best_balance_value = bert_time_df.loc[best_balance_idx, 'Balance']
        print(f"   • Mejor modelo: {best_model} (Balance: {best_balance_value:.2f})")
    
    print("─" * 50)
    
except Exception as e:
    print(f"❌ Error calculando balance: {e}")
    bert_time_df["Balance"] = 999999  # Valor por defecto en caso de error

# Mostrar DataFrame actualizado con balance
bert_time_df
````

````python
"""
VISUALIZACIÓN COMPARATIVA - TODOS LOS MODELOS
=============================================

Gráfico scatter comparativo de BertScore vs Tiempo de Inferencia para todos los modelos.
"""

print("📊 GENERANDO VISUALIZACIÓN COMPARATIVA - TODOS LOS MODELOS")
print("=" * 65)

try:
    # Identificar mejores modelos
    valid_data = bert_time_df[(bert_time_df['BertScore'] > 0) & (bert_time_df['Inference Time'] != float('inf'))]
    
    if not valid_data.empty:
        # Encontrar el mejor modelo por precisión y balance
        best_score_idx = valid_data['BertScore'].idxmax()
        best_balance_idx = valid_data['Balance'].idxmin()
        
        # Configurar el gráfico
        plt.figure(figsize=(12, 8))
        
        # Scatter plot principal - todos los modelos
        plt.scatter(valid_data['Inference Time'], valid_data['BertScore'], 
                   color='lightgray', s=80, alpha=0.7, label='Modelos IA', zorder=2)
        
        # Resaltar mejor por BertScore
        if best_score_idx in valid_data.index:
            plt.scatter(bert_time_df.loc[best_score_idx, 'Inference Time'], 
                       bert_time_df.loc[best_score_idx, 'BertScore'],
                       color='green', s=150, label=f"🏆 Mejor Precisión: {bert_time_df.loc[best_score_idx, 'Model']}", 
                       zorder=4, edgecolors='darkgreen', linewidth=2)
        
        # Resaltar mejor balance
        if best_balance_idx in valid_data.index and best_balance_idx != best_score_idx:
            plt.scatter(bert_time_df.loc[best_balance_idx, 'Inference Time'], 
                       bert_time_df.loc[best_balance_idx, 'BertScore'],
                       color='blue', s=150, label=f"⚖️ Mejor Balance: {bert_time_df.loc[best_balance_idx, 'Model']}", 
                       zorder=4, edgecolors='darkblue', linewidth=2)
        
        # Añadir etiquetas a cada punto
        for i in valid_data.index:
            model_name = bert_time_df.loc[i, 'Model']
            x_pos = bert_time_df.loc[i, 'Inference Time']
            y_pos = bert_time_df.loc[i, 'BertScore']
            
            # Ajustar posición del texto para evitar solapamiento
            plt.annotate(model_name, (x_pos, y_pos), 
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=9, alpha=0.8, 
                        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))
        
        # Configuración del gráfico
        plt.xlabel('⏱️ Tiempo de Inferencia (segundos)', fontsize=12, fontweight='bold')
        plt.ylabel('🎯 BertScore F1', fontsize=12, fontweight='bold')
        plt.title('📊 Análisis BertScore vs Velocidad - Preguntas No Estructuradas\n' +
                 'Comparativa de Modelos IA en Regulaciones Eléctricas', 
                 fontsize=14, fontweight='bold', pad=20)
        
        # Configuración de la leyenda y grilla
        plt.legend(loc='best', framealpha=0.9)
        plt.grid(True, alpha=0.3, linestyle='--')
        plt.tight_layout()
        
        # Añadir información estadística
        plt.figtext(0.02, 0.02, 
                   f'📊 Datos: {len(valid_data)} modelos válidos | ' +
                   f'BertScore promedio: {valid_data["BertScore"].mean():.3f} | ' +
                   f'Tiempo promedio: {valid_data["Inference Time"].mean():.2f}s',
                   fontsize=8, alpha=0.7)
        
        print("✅ Gráfico generado exitosamente")
        print(f"   • Modelos visualizados: {len(valid_data)}")
        print(f"   • Mejor precisión: {bert_time_df.loc[best_score_idx, 'Model']}")
        print(f"   • Mejor balance: {bert_time_df.loc[best_balance_idx, 'Model']}")
        
        plt.show()
        
    else:
        print("❌ No hay datos válidos para visualizar")
        print("   Verifica que los modelos tengan BertScore y tiempos válidos")
        
except Exception as e:
    print(f"❌ Error generando visualización: {e}")
    print("   Verifica que los datos estén correctamente cargados")
````

````python
"""
EVALUACIÓN BERTSCORE - PREGUNTAS ESTRUCTURADAS
=============================================

Análisis comparativo especializado para capacidades de procesamiento de datos estructurados.
"""

print("🔬 INICIANDO EVALUACIÓN BERTSCORE - DATOS ESTRUCTURADOS")
print("=" * 70)

# Reinicializar listas para este análisis
bertscore_by_model = []
times_by_model = []

# Procesamiento de datos estructurados con conversión a string
if not df_results.empty and "Respuesta Target" in df_results.columns:
    # Convertir respuestas objetivo a string para manejar datos numéricos
    targets = list(df_results["Respuesta Target"].astype(str))
    
    print(f"✅ Respuestas objetivo (estructuradas) cargadas: {len(targets)}")
    print(f"📊 Tipo de análisis: Datos tabulares y cálculos técnicos")
    
    print(f"\n🔄 PROCESANDO {len(models)} MODELOS PARA DATOS ESTRUCTURADOS...")
    print("─" * 60)
    
    # Evaluar cada modelo en datos estructurados
    for i, model in enumerate(models):
        print(f"🤖 Evaluando modelo: {model}")
        print(f"   📊 Tipo: Análisis de datos estructurados")
        
        try:
            # Extraer predicciones del modelo (asegurar conversión a string)
            predictions = list(df_results.iloc[:, i+2].astype(str))
            
            print(f"   🔍 Calculando BertScore para datos técnicos...")
            # Calcular BertScore (especialmente importante para precisión técnica)
            bert_score = get_bert_score(targets, predictions, scorer)
            bertscore_by_model.append(bert_score)
            
            # Extraer y procesar tiempos de inferencia
            print(f"   ⏱️  Analizando tiempos de procesamiento...")
            times = list(df_results.iloc[:, i+12])
            
            # Filtrar tiempos válidos (crítico para datos estructurados)
            valid_times = [
                float(x) for x in times
                if x != '' and x is not None and not isinstance(x, datetime)
            ]
            
            if valid_times:
                time_mean = statistics.mean(valid_times)
                times_by_model.append(time_mean)
                
                # Análisis adicional para datos estructurados
                time_std = statistics.stdev(valid_times) if len(valid_times) > 1 else 0
                print(f"   📈 Variabilidad temporal: ±{time_std:.3f}s")
            else:
                times_by_model.append(float('inf'))
                print(f"   ⚠️  Sin tiempos válidos disponibles")
            
            # Resultados del modelo con contexto técnico
            print(f"   ✅ BertScore técnico: {bert_score:.4f}")
            if bert_score > 0.85:
                print(f"       🟢 Excelente precisión técnica")
            elif bert_score > 0.70:
                print(f"       🟡 Precisión técnica aceptable") 
            else:
                print(f"       🔴 Precisión técnica insuficiente")
                
            print(f"   ⚡ Tiempo promedio: {time_mean:.3f}s")
            print(f"   📊 Mediciones válidas: {len(valid_times)}/{len(times)}")
            
        except Exception as e:
            print(f"   ❌ Error procesando {model}: {str(e)[:60]}...")
            bertscore_by_model.append(0.0)
            times_by_model.append(float('inf'))
        
        print("─" * 60)
    
    # Resumen especializado para datos estructurados
    valid_scores = [score for score in bertscore_by_model if score > 0]
    valid_times = [time for time in times_by_model if time != float('inf')]
    
    print(f"🎉 EVALUACIÓN DE DATOS ESTRUCTURADOS COMPLETADA")
    print(f"   • Modelos analizados: {len(bertscore_by_model)}")
    print(f"   • BertScores técnicos válidos: {len(valid_scores)}")
    print(f"   • Tiempos de procesamiento válidos: {len(valid_times)}")
    
    if valid_scores:
        avg_technical_score = statistics.mean(valid_scores)
        print(f"   • Precisión técnica promedio: {avg_technical_score:.4f}")
        print(f"   • Modelos con precisión excelente (>0.85): {sum(1 for s in valid_scores if s > 0.85)}")
        print(f"   • Modelos con precisión aceptable (0.70-0.85): {sum(1 for s in valid_scores if 0.70 <= s <= 0.85)}")
    
else:
    print("❌ No se pueden evaluar datos estructurados: datos insuficientes")
    bertscore_by_model = [0.0] * len(models)
    times_by_model = [float('inf')] * len(models)
````

````python
"""
CONSOLIDACIÓN DE RESULTADOS - DATOS ESTRUCTURADOS
================================================
"""

print("📊 CONSOLIDANDO RESULTADOS - DATOS ESTRUCTURADOS")
print("=" * 55)

# Crear DataFrame con resultados para datos estructurados
bert_time_df = pd.DataFrame({
    "Model": models,
    "BertScore": bertscore_by_model,
    "Inference Time": times_by_model
})

print(f"✅ DataFrame técnico creado: {bert_time_df.shape[0]} modelos")

# Estadísticas específicas para datos estructurados
valid_models = bert_time_df[(bert_time_df['BertScore'] > 0) & (bert_time_df['Inference Time'] != float('inf'))]

if not valid_models.empty:
    print(f"\n📈 ESTADÍSTICAS TÉCNICAS:")
    print(f"   • Modelos con capacidad técnica: {len(valid_models)}/{len(bert_time_df)}")
    print(f"   • Precisión técnica promedio: {valid_models['BertScore'].mean():.4f}")
    print(f"   • Mejor precisión técnica: {valid_models['BertScore'].max():.4f}")
    print(f"   • Tiempo de procesamiento promedio: {valid_models['Inference Time'].mean():.3f}s")

bert_time_df
````

````python
"""
CÁLCULO DE BALANCE TÉCNICO - DATOS ESTRUCTURADOS
===============================================
"""

print("⚖️  CALCULANDO BALANCE TÉCNICO PARA DATOS ESTRUCTURADOS")

try:
    bert_time_df["Balance"] = bert_time_df["Inference Time"] / bert_time_df["BertScore"]
    bert_time_df["Balance"] = bert_time_df["Balance"].replace([float('inf'), -float('inf')], 999999)
    
    valid_balance = bert_time_df[bert_time_df["Balance"] < 999999]
    
    if not valid_balance.empty:
        best_technical_balance = valid_balance['Balance'].idxmin()
        best_model = bert_time_df.loc[best_technical_balance, 'Model']
        print(f"✅ Mejor balance técnico: {best_model}")
        print(f"   Balance: {bert_time_df.loc[best_technical_balance, 'Balance']:.2f}")
    
except Exception as e:
    print(f"❌ Error en balance técnico: {e}")
    bert_time_df["Balance"] = 999999

bert_time_df
````
