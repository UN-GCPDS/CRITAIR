````python
"""
DEPENDENCIES CONFIGURATION - BERTSCORE EVALUATION
================================================

Import of all necessary libraries for comparative analysis
of AI models using BertScore as the main metric.
"""

# ===================================================================================================
# EVALUATION AND METRICS LIBRARIES
# ===================================================================================================

from bert_score import BERTScorer  # üéØ Main semantic evaluation metric

# ===================================================================================================
# DATA ANALYSIS LIBRARIES
# ===================================================================================================

import pandas as pd               # üìä Structured data manipulation and analysis
import statistics                 # üìà Statistical calculations (mean, deviation, etc.)
from datetime import datetime     # ‚è∞ Date and timestamp handling

# ===================================================================================================
# VISUALIZATION LIBRARIES
# ===================================================================================================

import matplotlib.pyplot as plt   # üìà Charts and visualizations creation

# ===================================================================================================
# FILE LIBRARIES
# ===================================================================================================

from openpyxl import load_workbook  # üìÑ Excel files reading with experimental results

print("‚úÖ All dependencies imported successfully")
print("üéØ System ready for BertScore evaluation")
````

````python
"""
BERTSCORE EVALUATION FUNCTION
============================

Implementation of the main function to calculate semantic similarity
between target responses and AI model predictions.
"""

def get_bert_score(targets, predictions, scorer):
    """
    Calculates semantic similarity between target sentences and predictions using BertScore.
    
    BertScore is a metric that evaluates generated text quality by comparing
    contextual token representations using pre-trained language models.
    
    Args:
        targets (List[str]): List of target sentences (ground truth)
        predictions (List[str]): List of sentences predicted by the model
        scorer (BERTScorer): Configured BertScore evaluator instance
    
    Returns:
        float: Average BertScore F1 score (0.0 - 1.0)
               - Values close to 1.0 indicate high semantic similarity
               - Values close to 0.0 indicate low semantic similarity
    
    Notes:
        - Uses F1 score that combines precision and recall
        - The mean() function calculates the average of all evaluated pairs
        - item() extracts the numerical value from PyTorch tensor
    """
    try:
        # Calculate precision, recall and F1 for each target-prediction pair
        precision, recall, f1_scores = scorer.score(targets, predictions)
        
        # Return the mean of F1 score as main metric
        bert_score_mean = f1_scores.mean().item()
        
        return bert_score_mean
        
    except Exception as e:
        print(f"‚ùå Error calculating BertScore: {e}")
        return 0.0  # Return 0 in case of error

print("‚úÖ get_bert_score function configured correctly")
print("üìä Metric: Average F1 Score for semantic evaluation")
````

````python
"""
CONFIGURACI√ìN DEL EVALUADOR BERTSCORE
====================================

Inicializaci√≥n del scorer BertScore con configuraci√≥n optimizada para espa√±ol.
"""

print("üîß Configuring BertScore evaluator...")

# Configure BertScore for Spanish evaluation with baseline rescaling
scorer = BERTScorer(
    lang="es",                    # üá™üá∏ Language: Spanish
    rescale_with_baseline=True    # üìä Baseline normalization for better comparisons
)

print("‚úÖ BERTScorer configured successfully")
print("üìã Configuration:")
print("   ‚Ä¢ Language: Spanish (es)")
print("   ‚Ä¢ Baseline rescaling: Enabled")
print("   ‚Ä¢ Metric: Contextualized F1 Score")
print("   ‚Ä¢ Underlying model: Based on pre-trained transformers")
````

````python
"""
EXPERIMENTAL DATA LOADING - UNSTRUCTURED QUESTIONS
=================================================

Loading dataset with experimental results for unstructured questions.
"""

print("üìä LOADING EXPERIMENTAL DATA - UNSTRUCTURED QUESTIONS")
print("=" * 70)

try:
    # ‚ö†Ô∏è PRIVACY NOTE: This path references experimental data
    # In a production environment, this data would be in a secure location
    file_path = "/content/drive/MyDrive/Thesis/Experimental SetUp/bitacora_chatbot_recomendaci√≥n_wo_think.xlsx"
    
    print("üìÇ Loading experimental results file...")
    print(f"   File: {file_path.split('/')[-1]}")
    print(f"   Sheet: No structure questions")
    
    # Load unstructured questions data
    df_results = pd.read_excel(
        file_path,
        sheet_name="No structure questions"
    )
    
    print(f"‚úÖ Data loaded successfully")
    print(f"üìã Dataset dimensions: {df_results.shape}")
    print(f"   ‚Ä¢ Rows (questions): {df_results.shape[0]}")
    print(f"   ‚Ä¢ Columns (models + metadata): {df_results.shape[1]}")
    
    # Show dataset information
    if not df_results.empty:
        print(f"\nüìä DATASET INFORMATION:")
        print(f"   ‚Ä¢ Questions evaluated: {len(df_results)}")
        print(f"   ‚Ä¢ Available columns: {', '.join(df_results.columns[:5])}...")
        
        # Verify target responses column
        if "Respuesta Target" in df_results.columns:
            print(f"   ‚Ä¢ Target responses available: ‚úÖ")
        else:
            print(f"   ‚Ä¢ Target responses available: ‚ùå")
    
    # Show dataset preview
    df_results
    
except FileNotFoundError:
    print("‚ùå Error: Data file not found")
    print("   Verify that the Excel file is available at the specified path")
    df_results = pd.DataFrame()  # Empty dataset to avoid errors
    
except Exception as e:
    print(f"‚ùå Error loading data: {e}")
    df_results = pd.DataFrame()
````

````python
"""
BERTSCORE EVALUATION - UNSTRUCTURED QUESTIONS
============================================

Comparative analysis of all AI models for unstructured questions
using BertScore as semantic similarity metric.
"""

print("üî¨ STARTING BERTSCORE EVALUATION - UNSTRUCTURED DATA")
print("=" * 70)

# Initialize lists to store results
bertscore_by_model = []
times_by_model = []

# Extract target responses (ground truth)
if not df_results.empty and "Respuesta Target" in df_results.columns:
    targets = list(df_results["Respuesta Target"])
    print(f"‚úÖ Target responses loaded: {len(targets)}")
    
    print(f"\nüîÑ PROCESSING {len(models)} MODELS...")
    print("‚îÄ" * 50)
    
    # Evaluate each model
    for i, model in enumerate(models):
        print(f"ü§ñ Evaluating: {model}")
        
        try:
            # Extract model predictions (column i+2 because first ones are metadata)
            predictions = list(df_results.iloc[:, i+2])
            
            # Calculate BertScore
            print(f"   üìä Calculating BertScore...")
            bert_score = get_bert_score(targets, predictions, scorer)
            bertscore_by_model.append(bert_score)
            
            # Extract inference times (column i+12)
            print(f"   ‚è±Ô∏è  Calculating average times...")
            times = list(df_results.iloc[:, i+12])
            
            # Clean and process times (remove invalid values)
            valid_times = [
                float(x) for x in times
                if x != '' and x is not None and not isinstance(x, datetime)
            ]
            
            if valid_times:
                time_mean = statistics.mean(valid_times)
                times_by_model.append(time_mean)
            else:
                times_by_model.append(float('inf'))  # Infinite time if no valid data
                
            # Show model results
            print(f"   ‚úÖ BertScore: {bert_score:.4f}")
            print(f"   ‚ö° Average time: {time_mean:.3f}s")
            print(f"   üìä Valid measurements: {len(valid_times)}/{len(times)}")
            
        except Exception as e:
            print(f"   ‚ùå Error processing {model}: {str(e)[:50]}...")
            bertscore_by_model.append(0.0)
            times_by_model.append(float('inf'))
        
        print("‚îÄ" * 50)
    
    print(f"üéâ EVALUATION COMPLETED")
    print(f"   ‚Ä¢ Models evaluated: {len(bertscore_by_model)}")
    print(f"   ‚Ä¢ BertScores calculated: {sum(1 for score in bertscore_by_model if score > 0)}")
    print(f"   ‚Ä¢ Valid times: {sum(1 for time in times_by_model if time != float('inf'))}")
    
else:
    print("‚ùå Cannot evaluate models: insufficient data")
    bertscore_by_model = [0.0] * len(models)
    times_by_model = [float('inf')] * len(models)
````

````python
"""
RESULTS CONSOLIDATION - UNSTRUCTURED DATA
=========================================

Creation of DataFrame with consolidated BertScore and time results.
"""

print("üìä CONSOLIDATING RESULTS - UNSTRUCTURED QUESTIONS")
print("=" * 60)

# Crear DataFrame con resultados
bert_time_df = pd.DataFrame({
    "Model": models,
    "BertScore": bertscore_by_model,
    "Inference Time": times_by_model
})

print("‚úÖ DataFrame de resultados creado")
print(f"üìã Estructura: {bert_time_df.shape[0]} modelos √ó {bert_time_df.shape[1]} m√©tricas")

# Mostrar estad√≠sticas r√°pidas
if not bert_time_df.empty:
    valid_scores = bert_time_df[bert_time_df['BertScore'] > 0]
    valid_times = bert_time_df[bert_time_df['Inference Time'] != float('inf')]
    
    print(f"\nüìà ESTAD√çSTICAS GENERALES:")
    print(f"   ‚Ä¢ Modelos con BertScore v√°lido: {len(valid_scores)}/{len(bert_time_df)}")
    print(f"   ‚Ä¢ Modelos con tiempo v√°lido: {len(valid_times)}/{len(bert_time_df)}")
    
    if len(valid_scores) > 0:
        print(f"   ‚Ä¢ BertScore promedio: {valid_scores['BertScore'].mean():.4f}")
        print(f"   ‚Ä¢ Mejor BertScore: {valid_scores['BertScore'].max():.4f}")
        print(f"   ‚Ä¢ Peor BertScore: {valid_scores['BertScore'].min():.4f}")
    
    if len(valid_times) > 0:
        print(f"   ‚Ä¢ Tiempo promedio: {valid_times['Inference Time'].mean():.3f}s")
        print(f"   ‚Ä¢ Modelo m√°s r√°pido: {valid_times['Inference Time'].min():.3f}s")
        print(f"   ‚Ä¢ Modelo m√°s lento: {valid_times['Inference Time'].max():.3f}s")

# Mostrar el DataFrame completo
bert_time_df
````

````python
"""
C√ÅLCULO DE BALANCE CALIDAD/VELOCIDAD
===================================

C√°lculo de la m√©trica de balance que combina BertScore y tiempo de inferencia.
"""

print("‚öñÔ∏è  CALCULANDO BALANCE CALIDAD/VELOCIDAD")
print("=" * 50)

# Calcular balance: Tiempo/BertScore (menor es mejor)
# - Valores bajos indican buen balance (alta calidad, tiempo razonable)
# - Valores altos indican mal balance (baja calidad o tiempo excesivo)
try:
    bert_time_df["Balance"] = bert_time_df["Inference Time"] / bert_time_df["BertScore"]
    
    # Manejar casos especiales
    bert_time_df["Balance"] = bert_time_df["Balance"].replace([float('inf'), -float('inf')], 999999)
    
    print("‚úÖ M√©trica de balance calculada")
    print("\nüìä INTERPRETACI√ìN DEL BALANCE:")
    print("   ‚Ä¢ Valores bajos (< 10):  Excelente balance calidad/velocidad")
    print("   ‚Ä¢ Valores medios (10-50): Balance aceptable") 
    print("   ‚Ä¢ Valores altos (> 50):   Balance deficiente")
    
    # Mostrar estad√≠sticas del balance
    valid_balance = bert_time_df[bert_time_df["Balance"] < 999999]
    
    if not valid_balance.empty:
        print(f"\nüìà ESTAD√çSTICAS DE BALANCE:")
        print(f"   ‚Ä¢ Modelos con balance v√°lido: {len(valid_balance)}")
        print(f"   ‚Ä¢ Mejor balance: {valid_balance['Balance'].min():.2f}")
        print(f"   ‚Ä¢ Peor balance: {valid_balance['Balance'].max():.2f}")
        print(f"   ‚Ä¢ Balance promedio: {valid_balance['Balance'].mean():.2f}")
        
        # Identificar mejor modelo por balance
        best_balance_idx = valid_balance['Balance'].idxmin()
        best_model = bert_time_df.loc[best_balance_idx, 'Model']
        best_balance_value = bert_time_df.loc[best_balance_idx, 'Balance']
        print(f"   ‚Ä¢ Mejor modelo: {best_model} (Balance: {best_balance_value:.2f})")
    
    print("‚îÄ" * 50)
    
except Exception as e:
    print(f"‚ùå Error calculando balance: {e}")
    bert_time_df["Balance"] = 999999  # Valor por defecto en caso de error

# Mostrar DataFrame actualizado con balance
bert_time_df
````

````python
"""
VISUALIZACI√ìN COMPARATIVA - TODOS LOS MODELOS
=============================================

Gr√°fico scatter comparativo de BertScore vs Tiempo de Inferencia para todos los modelos.
"""

print("üìä GENERANDO VISUALIZACI√ìN COMPARATIVA - TODOS LOS MODELOS")
print("=" * 65)

try:
    # Identificar mejores modelos
    valid_data = bert_time_df[(bert_time_df['BertScore'] > 0) & (bert_time_df['Inference Time'] != float('inf'))]
    
    if not valid_data.empty:
        # Encontrar el mejor modelo por precisi√≥n y balance
        best_score_idx = valid_data['BertScore'].idxmax()
        best_balance_idx = valid_data['Balance'].idxmin()
        
        # Configurar el gr√°fico
        plt.figure(figsize=(12, 8))
        
        # Scatter plot principal - todos los modelos
        plt.scatter(valid_data['Inference Time'], valid_data['BertScore'], 
                   color='lightgray', s=80, alpha=0.7, label='Modelos IA', zorder=2)
        
        # Resaltar mejor por BertScore
        if best_score_idx in valid_data.index:
            plt.scatter(bert_time_df.loc[best_score_idx, 'Inference Time'], 
                       bert_time_df.loc[best_score_idx, 'BertScore'],
                       color='green', s=150, label=f"üèÜ Mejor Precisi√≥n: {bert_time_df.loc[best_score_idx, 'Model']}", 
                       zorder=4, edgecolors='darkgreen', linewidth=2)
        
        # Resaltar mejor balance
        if best_balance_idx in valid_data.index and best_balance_idx != best_score_idx:
            plt.scatter(bert_time_df.loc[best_balance_idx, 'Inference Time'], 
                       bert_time_df.loc[best_balance_idx, 'BertScore'],
                       color='blue', s=150, label=f"‚öñÔ∏è Mejor Balance: {bert_time_df.loc[best_balance_idx, 'Model']}", 
                       zorder=4, edgecolors='darkblue', linewidth=2)
        
        # A√±adir etiquetas a cada punto
        for i in valid_data.index:
            model_name = bert_time_df.loc[i, 'Model']
            x_pos = bert_time_df.loc[i, 'Inference Time']
            y_pos = bert_time_df.loc[i, 'BertScore']
            
            # Ajustar posici√≥n del texto para evitar solapamiento
            plt.annotate(model_name, (x_pos, y_pos), 
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=9, alpha=0.8, 
                        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))
        
        # Configuraci√≥n del gr√°fico
        plt.xlabel('‚è±Ô∏è Tiempo de Inferencia (segundos)', fontsize=12, fontweight='bold')
        plt.ylabel('üéØ BertScore F1', fontsize=12, fontweight='bold')
        plt.title('üìä An√°lisis BertScore vs Velocidad - Preguntas No Estructuradas\n' +
                 'Comparativa de Modelos IA en Regulaciones El√©ctricas', 
                 fontsize=14, fontweight='bold', pad=20)
        
        # Configuraci√≥n de la leyenda y grilla
        plt.legend(loc='best', framealpha=0.9)
        plt.grid(True, alpha=0.3, linestyle='--')
        plt.tight_layout()
        
        # A√±adir informaci√≥n estad√≠stica
        plt.figtext(0.02, 0.02, 
                   f'üìä Datos: {len(valid_data)} modelos v√°lidos | ' +
                   f'BertScore promedio: {valid_data["BertScore"].mean():.3f} | ' +
                   f'Tiempo promedio: {valid_data["Inference Time"].mean():.2f}s',
                   fontsize=8, alpha=0.7)
        
        print("‚úÖ Gr√°fico generado exitosamente")
        print(f"   ‚Ä¢ Modelos visualizados: {len(valid_data)}")
        print(f"   ‚Ä¢ Mejor precisi√≥n: {bert_time_df.loc[best_score_idx, 'Model']}")
        print(f"   ‚Ä¢ Mejor balance: {bert_time_df.loc[best_balance_idx, 'Model']}")
        
        plt.show()
        
    else:
        print("‚ùå No hay datos v√°lidos para visualizar")
        print("   Verifica que los modelos tengan BertScore y tiempos v√°lidos")
        
except Exception as e:
    print(f"‚ùå Error generando visualizaci√≥n: {e}")
    print("   Verifica que los datos est√©n correctamente cargados")
````

````python
"""
EVALUACI√ìN BERTSCORE - PREGUNTAS ESTRUCTURADAS
=============================================

An√°lisis comparativo especializado para capacidades de procesamiento de datos estructurados.
"""

print("üî¨ INICIANDO EVALUACI√ìN BERTSCORE - DATOS ESTRUCTURADOS")
print("=" * 70)

# Reinicializar listas para este an√°lisis
bertscore_by_model = []
times_by_model = []

# Procesamiento de datos estructurados con conversi√≥n a string
if not df_results.empty and "Respuesta Target" in df_results.columns:
    # Convertir respuestas objetivo a string para manejar datos num√©ricos
    targets = list(df_results["Respuesta Target"].astype(str))
    
    print(f"‚úÖ Respuestas objetivo (estructuradas) cargadas: {len(targets)}")
    print(f"üìä Tipo de an√°lisis: Datos tabulares y c√°lculos t√©cnicos")
    
    print(f"\nüîÑ PROCESANDO {len(models)} MODELOS PARA DATOS ESTRUCTURADOS...")
    print("‚îÄ" * 60)
    
    # Evaluar cada modelo en datos estructurados
    for i, model in enumerate(models):
        print(f"ü§ñ Evaluando modelo: {model}")
        print(f"   üìä Tipo: An√°lisis de datos estructurados")
        
        try:
            # Extraer predicciones del modelo (asegurar conversi√≥n a string)
            predictions = list(df_results.iloc[:, i+2].astype(str))
            
            print(f"   üîç Calculando BertScore para datos t√©cnicos...")
            # Calcular BertScore (especialmente importante para precisi√≥n t√©cnica)
            bert_score = get_bert_score(targets, predictions, scorer)
            bertscore_by_model.append(bert_score)
            
            # Extraer y procesar tiempos de inferencia
            print(f"   ‚è±Ô∏è  Analizando tiempos de procesamiento...")
            times = list(df_results.iloc[:, i+12])
            
            # Filtrar tiempos v√°lidos (cr√≠tico para datos estructurados)
            valid_times = [
                float(x) for x in times
                if x != '' and x is not None and not isinstance(x, datetime)
            ]
            
            if valid_times:
                time_mean = statistics.mean(valid_times)
                times_by_model.append(time_mean)
                
                # An√°lisis adicional para datos estructurados
                time_std = statistics.stdev(valid_times) if len(valid_times) > 1 else 0
                print(f"   üìà Variabilidad temporal: ¬±{time_std:.3f}s")
            else:
                times_by_model.append(float('inf'))
                print(f"   ‚ö†Ô∏è  Sin tiempos v√°lidos disponibles")
            
            # Resultados del modelo con contexto t√©cnico
            print(f"   ‚úÖ BertScore t√©cnico: {bert_score:.4f}")
            if bert_score > 0.85:
                print(f"       üü¢ Excelente precisi√≥n t√©cnica")
            elif bert_score > 0.70:
                print(f"       üü° Precisi√≥n t√©cnica aceptable") 
            else:
                print(f"       üî¥ Precisi√≥n t√©cnica insuficiente")
                
            print(f"   ‚ö° Tiempo promedio: {time_mean:.3f}s")
            print(f"   üìä Mediciones v√°lidas: {len(valid_times)}/{len(times)}")
            
        except Exception as e:
            print(f"   ‚ùå Error procesando {model}: {str(e)[:60]}...")
            bertscore_by_model.append(0.0)
            times_by_model.append(float('inf'))
        
        print("‚îÄ" * 60)
    
    # Resumen especializado para datos estructurados
    valid_scores = [score for score in bertscore_by_model if score > 0]
    valid_times = [time for time in times_by_model if time != float('inf')]
    
    print(f"üéâ EVALUACI√ìN DE DATOS ESTRUCTURADOS COMPLETADA")
    print(f"   ‚Ä¢ Modelos analizados: {len(bertscore_by_model)}")
    print(f"   ‚Ä¢ BertScores t√©cnicos v√°lidos: {len(valid_scores)}")
    print(f"   ‚Ä¢ Tiempos de procesamiento v√°lidos: {len(valid_times)}")
    
    if valid_scores:
        avg_technical_score = statistics.mean(valid_scores)
        print(f"   ‚Ä¢ Precisi√≥n t√©cnica promedio: {avg_technical_score:.4f}")
        print(f"   ‚Ä¢ Modelos con precisi√≥n excelente (>0.85): {sum(1 for s in valid_scores if s > 0.85)}")
        print(f"   ‚Ä¢ Modelos con precisi√≥n aceptable (0.70-0.85): {sum(1 for s in valid_scores if 0.70 <= s <= 0.85)}")
    
else:
    print("‚ùå No se pueden evaluar datos estructurados: datos insuficientes")
    bertscore_by_model = [0.0] * len(models)
    times_by_model = [float('inf')] * len(models)
````
