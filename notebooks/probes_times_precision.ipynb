````python
"""
PROCESAMIENTO PRINCIPAL: AN√ÅLISIS NO ESTRUCTURADO
================================================

Este bloque ejecuta el an√°lisis completo de consultas no estructuradas utilizando
un agente LangChain especializado con acceso a herramientas de documentos normativos.
"""

# ===================================================================================================
# CONFIGURACI√ìN DEL EXPERIMENTO
# ===================================================================================================

# Configuraci√≥n del modelo a evaluar
TARGET_MODEL = "gemini-2.0-flash-001"
EXPERIMENT_NAME = f"unstructured_analysis_{TARGET_MODEL.replace('-', '_').replace('.', '_')}"

# Estructuras para almacenar resultados
all_responses = []
all_times = []
processing_details = []

print(f"üöÄ INICIANDO EXPERIMENTO: {EXPERIMENT_NAME}")
print(f"üìä Modelo objetivo: {TARGET_MODEL}")
print(f"‚ùì Preguntas a procesar: {len(questions)}")
print("=" * 80)

# ===================================================================================================
# CONFIGURACI√ìN DE HERRAMIENTAS Y AGENTE
# ===================================================================================================

# Lista completa de herramientas especializadas
AVAILABLE_TOOLS = [
    capitulo_1, capitulo_2, capitulo_3, capitulo_4,
    resolucion_40117, normativa_apoyos, normativa_protecciones, 
    normativa_aisladores, redes_aereas_media_tension, 
    codigo_electrico_colombiano, requisitos_redes_aereas, 
    retie, eventos_transformadores, eventos_transformadores_plots
]

print(f"üîß Herramientas disponibles: {len(AVAILABLE_TOOLS)}")

# ===================================================================================================
# PROCESAMIENTO ITERATIVO DE PREGUNTAS
# ===================================================================================================

for question_idx, query in enumerate(questions, 1):
    print(f"\nüìù PROCESANDO PREGUNTA {question_idx}/{len(questions)}")
    print(f"‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")
    
    # Preview de la pregunta
    preview = query[:100] + "..." if len(query) > 100 else query
    print(f"üîç Pregunta: {preview}")
    
    # Inicializar medici√≥n de tiempo
    start_time = time.time()
    
    # Configuraci√≥n del sistema de archivos de iteraci√≥n
    iteration_file = "number_iteration.pkl"
    with open(iteration_file, "w") as f:
        f.write("0")  # Resetear contador de iteraci√≥n
    
    # Configuraci√≥n del agente LLM
    try:
        llm_agent = ChatOpenAI(
            temperature=0, 
            model="gpt-3.5-turbo",
            http_client=httpx.Client(verify=False)
        )
        print("üîó Cliente HTTP est√°ndar configurado")
    except Exception as e:
        print(f"‚ö†Ô∏è  Fallback a cliente as√≠ncrono: {e}")
        llm_agent = ChatOpenAI(
            temperature=0, 
            model="gpt-3.5-turbo",
            http_client=httpx.AsyncClient(verify=False)
        )
    
    # Configuraci√≥n del prompt especializado
    prompt = hub.pull("hwchase17/openai-functions-agent")
    prompt.messages[0].prompt.template = """
    Eres un asistente t√©cnico especializado en normativas el√©ctricas colombianas con una conversaci√≥n 
    con un ingeniero el√©ctrico. Tu funci√≥n es responder preguntas t√©cnicas precisas utilizando la 
    informaci√≥n de los documentos normativos disponibles.
    
    INSTRUCCIONES ESPEC√çFICAS:
    1. Dada una pregunta t√©cnica, selecciona la herramienta m√°s apropiada que contenga la informaci√≥n relevante
    2. Si la pregunta no est√° relacionada con ninguna normativa disponible, explica qu√© tipos de consultas 
       puedes responder bas√°ndote en las herramientas disponibles
    3. Proporciona respuestas t√©cnicamente precisas y cita las secciones normativas cuando sea posible
    4. Si necesitas m√∫ltiples herramientas, √∫salas de manera coordinada para dar una respuesta completa
    
    HERRAMIENTAS DISPONIBLES: RETIE, NTC2050, Resoluciones CREG, normas t√©cnicas especializadas
    """
    
    # Creaci√≥n y configuraci√≥n del agente
    agent = create_openai_functions_agent(llm_agent, AVAILABLE_TOOLS, prompt)
    agent_executor = AgentExecutor(
        agent=agent, 
        tools=AVAILABLE_TOOLS, 
        verbose=True,
        max_iterations=10,
        handle_parsing_errors=True
    )
    
    # Mapeo de modelos para compatibilidad del sistema
    model_mapping = {
        "llama3.2": "llama2",
        "llama3.1": "llama1"
    }
    processed_model = model_mapping.get(TARGET_MODEL, TARGET_MODEL)
    
    try:
        print("ü§ñ Ejecutando consulta con agente...")
        execution_result = agent_executor.invoke({
            "input": [query, processed_model, str(question_idx)]
        })
        print("‚úÖ Consulta ejecutada exitosamente")
        
    except Exception as e:
        print(f"‚ùå Error en ejecuci√≥n del agente: {e}")
        execution_result = {"output": f"Error en procesamiento: {str(e)[:200]}..."}
    
    # Procesamiento de resultados seg√∫n estado del sistema
    try:
        with open(iteration_file, "r") as f:
            iteration_count = int(f.read().strip())
    except (FileNotFoundError, ValueError):
        iteration_count = 0
    
    # Procesar respuesta seg√∫n el estado del sistema
    if iteration_count == 0:
        final_response = execution_result["output"]
        has_visualization = False
    elif iteration_count == 3:
        try:
            with open("answer.pkl", 'rb') as f:
                final_response = pickle.load(f)
            has_visualization = True
        except FileNotFoundError:
            final_response = execution_result["output"]
            has_visualization = False
    elif iteration_count == 4:
        try:
            with open("answer.pkl", 'rb') as f:
                final_response = pickle.load(f)
            has_visualization = False
        except FileNotFoundError:
            final_response = execution_result["output"]
            has_visualization = False
    else:
        try:
            with open("answer.pkl", 'rb') as f:
                final_response = pickle.load(f)
            has_visualization = False
        except FileNotFoundError:
            final_response = execution_result["output"]
            has_visualization = False
    
    # Finalizaci√≥n y m√©tricas
    end_time = time.time()
    processing_time = end_time - start_time
    
    # Almacenar resultados
    all_responses.append(final_response)
    all_times.append(processing_time)
    
    # Almacenar detalles adicionales
    processing_details.append({
        'question_id': question_idx,
        'model': TARGET_MODEL,
        'processing_time': processing_time,
        'iteration_count': iteration_count,
        'has_visualization': has_visualization,
        'response_length': len(str(final_response)),
        'timestamp': time.strftime('%Y-%m-%d %H:%M:%S')
    })
    
    # Mostrar resumen de la pregunta
    print(f"‚è±Ô∏è  Tiempo de procesamiento: {processing_time:.2f}s")
    print(f"üîÑ Iteraciones del sistema: {iteration_count}")
    print(f"üìä Longitud de respuesta: {len(str(final_response))} caracteres")
    print(f"üé® Incluye visualizaci√≥n: {'S√≠' if has_visualization else 'No'}")
    
    # Preview de la respuesta
    response_preview = str(final_response)[:200] + "..." if len(str(final_response)) > 200 else str(final_response)
    print(f"üìã Respuesta: {response_preview}")
    print("‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")

# Resumen final del experimento
print(f"\nüéâ EXPERIMENTO COMPLETADO: {EXPERIMENT_NAME}")
print("=" * 80)
print(f"üìä ESTAD√çSTICAS FINALES:")
print(f"   ‚Ä¢ Preguntas procesadas: {len(all_responses)}")
print(f"   ‚Ä¢ Tiempo total: {sum(all_times):.2f} segundos")
print(f"   ‚Ä¢ Tiempo promedio: {sum(all_times)/len(all_times):.2f} segundos")
print(f"   ‚Ä¢ Tiempo m√≠nimo: {min(all_times):.2f} segundos")
print(f"   ‚Ä¢ Tiempo m√°ximo: {max(all_times):.2f} segundos")
print(f"   ‚Ä¢ Modelo evaluado: {TARGET_MODEL}")
print("=" * 80)
````

````python
"""
SECCI√ìN 3: AN√ÅLISIS DE PRECISI√ìN Y TIEMPOS - MODELOS DE RECOMENDACI√ìN
====================================================================

Esta secci√≥n ejecuta un an√°lisis especializado de precisi√≥n y tiempos de respuesta
para preguntas contextualizadas del RETIE utilizando m√∫ltiples modelos IA.
"""

# ===================================================================================================
# CARGA DE CONFIGURACI√ìN Y DATOS
# ===================================================================================================

print("üöÄ INICIANDO AN√ÅLISIS DE PRECISI√ìN Y TIEMPOS")
print("=" * 80)

try:
    # Cargar modelos y preguntas desde archivos modulares
    print("üìä Cargando configuraci√≥n del experimento...")
    
    # Obtener lista de modelos IA disponibles
    ai_models = get_ai_models()
    print(f"‚úÖ Modelos IA cargados: {len(ai_models)}")
    
    # Obtener preguntas contextualizadas del RETIE
    retie_questions = get_recommendation_questions()
    print(f"‚úÖ Preguntas RETIE cargadas: {len(retie_questions)}")
    
except Exception as e:
    print(f"‚ùå Error cargando configuraci√≥n: {e}")
    print("   Verifica que los m√≥dulos de preguntas est√©n disponibles")
    ai_models = []
    retie_questions = []

if not ai_models or not retie_questions:
    print("‚ö†Ô∏è  No se puede continuar sin modelos o preguntas")
    print("   Verifica la configuraci√≥n de los m√≥dulos")
else:
    # ===================================================================================================
    # VISUALIZACI√ìN DE CONFIGURACI√ìN
    # ===================================================================================================
    
    print(f"\nüìã CONFIGURACI√ìN DEL EXPERIMENTO:")
    print("‚îÄ" * 50)
    
    # Mostrar modelos disponibles
    print(f"ü§ñ MODELOS IA DISPONIBLES ({len(ai_models)}):")
    for i, model in enumerate(ai_models, 1):
        print(f"   {i:2d}. {model}")
    
    # Mostrar preview de preguntas
    print(f"\n‚ùì PREGUNTAS CONTEXTUALIZADAS DEL RETIE ({len(retie_questions)}):")
    for i, question in enumerate(retie_questions, 1):
        # Crear preview legible limitando la longitud
        preview = question[:80] + "..." if len(question) > 80 else question
        # Limpiar saltos de l√≠nea para mejor visualizaci√≥n
        preview = preview.replace('\n', ' ').replace('  ', ' ')
        print(f"   {i:2d}. {preview}")
    
    # ===================================================================================================
    # CONFIGURACI√ìN DEL EXPERIMENTO
    # ===================================================================================================
    
    print(f"\nüî¨ CONFIGURACI√ìN EXPERIMENTAL:")
    print("‚îÄ" * 40)
    
    # Modelos seleccionados para este experimento (seg√∫n l√≥gica original)
    SELECTED_MODEL_INDICES = [5, 6, 8, 9]  # √çndices espec√≠ficos del experimento original
    RESULTS_DIRECTORY = "times_vs_tokens"
    
    # Crear directorio de resultados
    os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
    
    print(f"   ‚Ä¢ Modelos seleccionados: {len(SELECTED_MODEL_INDICES)} de {len(ai_models)}")
    print(f"   ‚Ä¢ Preguntas por modelo: {len(retie_questions)}")
    print(f"   ‚Ä¢ Total de consultas: {len(SELECTED_MODEL_INDICES) * len(retie_questions)}")
    print(f"   ‚Ä¢ Directorio de resultados: {RESULTS_DIRECTORY}/")
    
    # ===================================================================================================
    # PROCESAMIENTO POR MODELO SELECCIONADO
    # ===================================================================================================
    
    print(f"\n‚ö° INICIANDO AN√ÅLISIS DE PRECISI√ìN Y TIEMPOS")
    print("=" * 80)
    
    experiment_results = {}
    total_queries = 0
    total_experiment_time = 0
    
    for model_index in SELECTED_MODEL_INDICES:
        if model_index >= len(ai_models):
            print(f"‚ö†Ô∏è  √çndice de modelo fuera de rango: {model_index}")
            continue
            
        model_name = ai_models[model_index]
        print(f"\nü§ñ PROCESANDO MODELO: {model_name}")
        print("‚îÄ" * 60)
        
        model_start_time = time.time()
        
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        # CONFIGURACI√ìN DEL MODELO ESPEC√çFICO
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        
        try:
            # Configurar modelo seg√∫n el tipo (basado en l√≥gica original)
            if model_index < 2:
                # Modelos OpenAI
                chat_model = ChatOpenAI(temperature=0, model=model_name)
                model_type = "OpenAI"
            elif 2 <= model_index < 4:
                # Modelos Google
                chat_model = ChatGoogleGenerativeAI(temperature=0, model=model_name)
                model_type = "Google"
            else:
                # Modelos Ollama
                chat_model = ChatOllama(model=model_name, temperature=0)
                model_type = "Ollama"
                
            print(f"‚úÖ Modelo configurado: {model_name} ({model_type})")
        
        except Exception as e:
            print(f"‚ùå Error configurando modelo {model_name}: {e}")
            continue
        
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        # PROCESAMIENTO DE PREGUNTAS
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        
        model_times = []
        model_responses = []
        
        print(f"üìù Procesando {len(retie_questions)} preguntas...")
        
        for question_idx, question_text in enumerate(retie_questions, 1):
            print(f"   Pregunta {question_idx}/{len(retie_questions)}", end=" ")
            
            try:
                # Medir tiempo de respuesta
                query_start = time.time()
                
                # Ejecutar consulta
                response = chat_model([HumanMessage(content=question_text)])
                
                query_end = time.time()
                response_time = query_end - query_start
                
                # Almacenar resultados
                model_times.append(response_time)
                model_responses.append(response)
                
                print(f"‚úÖ {response_time:.3f}s")
                
            except Exception as e:
                print(f"‚ùå Error: {str(e)[:50]}...")
                model_times.append(float('inf'))  # Marcar como error
                model_responses.append(f"Error: {str(e)[:100]}")
        
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        # AN√ÅLISIS Y GUARDADO DE RESULTADOS
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        
        model_end_time = time.time()
        model_total_time = model_end_time - model_start_time
        
        # Filtrar tiempos v√°lidos para estad√≠sticas
        valid_times = [t for t in model_times if t != float('inf')]
        
        # Calcular estad√≠sticas
        if valid_times:
            stats = {
                'total_queries': len(retie_questions),
                'successful_queries': len(valid_times),
                'failed_queries': len(retie_questions) - len(valid_times),
                'total_time': model_total_time,
                'avg_response_time': np.mean(valid_times),
                'min_response_time': min(valid_times),
                'max_response_time': max(valid_times),
                'std_response_time': np.std(valid_times),
                'success_rate': len(valid_times) / len(retie_questions) * 100
            }
        else:
            stats = {
                'total_queries': len(retie_questions),
                'successful_queries': 0,
                'failed_queries': len(retie_questions),
                'total_time': model_total_time,
                'success_rate': 0
            }
        
        # Almacenar en resultados globales
        experiment_results[model_name] = {
            'times': model_times,
            'responses': model_responses,
            'statistics': stats,
            'model_type': model_type
        }
        
        # Guardar tiempos en archivo de texto
        output_file = f"{RESULTS_DIRECTORY}/{model_name}.txt"
        try:
            with open(output_file, "w", encoding="utf-8") as f:
                for time_value in model_times:
                    if time_value != float('inf'):
                        f.write(f"{time_value}\n")
                    else:
                        f.write("ERROR\n")
            
            print(f"üíæ Resultados guardados: {output_file}")
        
        except Exception as e:
            print(f"‚ùå Error guardando {output_file}: {e}")
        
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        # MOSTRAR ESTAD√çSTICAS DEL MODELO
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        
        print(f"\nüìä ESTAD√çSTICAS - {model_name}:")
        print(f"   ‚Ä¢ Consultas exitosas: {stats['successful_queries']}/{stats['total_queries']}")
        print(f"   ‚Ä¢ Tasa de √©xito: {stats['success_rate']:.1f}%")
        print(f"   ‚Ä¢ Tiempo total: {stats['total_time']:.2f}s")
        
        if stats['successful_queries'] > 0:
            print(f"   ‚Ä¢ Tiempo promedio: {stats['avg_response_time']:.3f}s")
            print(f"   ‚Ä¢ Tiempo m√≠nimo: {stats['min_response_time']:.3f}s")
            print(f"   ‚Ä¢ Tiempo m√°ximo: {stats['max_response_time']:.3f}s")
            print(f"   ‚Ä¢ Desviaci√≥n est√°ndar: {stats['std_response_time']:.3f}s")
        
        total_queries += stats['total_queries']
        total_experiment_time += stats['total_time']
        
        print("‚îÄ" * 60)
    
    # ===================================================================================================
    # RESUMEN FINAL DEL EXPERIMENTO
    # ===================================================================================================
    
    print(f"\nüéâ AN√ÅLISIS DE PRECISI√ìN Y TIEMPOS COMPLETADO")
    print("=" * 80)
    
    successful_models = [name for name, data in experiment_results.items() 
                        if data['statistics']['successful_queries'] > 0]
    
    print(f"üìä RESUMEN FINAL:")
    print(f"   ‚Ä¢ Modelos procesados: {len(experiment_results)}")
    print(f"   ‚Ä¢ Modelos exitosos: {len(successful_models)}")
    print(f"   ‚Ä¢ Total de consultas: {total_queries}")
    print(f"   ‚Ä¢ Tiempo total del experimento: {total_experiment_time:.2f}s")
    
    if successful_models:
        # Calcular promedios globales
        all_success_rates = [data['statistics']['success_rate'] 
                           for data in experiment_results.values()]
        avg_success_rate = np.mean(all_success_rates)
        
        all_avg_times = [data['statistics']['avg_response_time'] 
                        for data in experiment_results.values() 
                        if data['statistics']['successful_queries'] > 0]
        
        if all_avg_times:
            global_avg_time = np.mean(all_avg_times)
            print(f"   ‚Ä¢ Tasa de √©xito promedio: {avg_success_rate:.1f}%")
            print(f"   ‚Ä¢ Tiempo promedio global: {global_avg_time:.3f}s")
    
    print(f"\nüìÅ ARCHIVOS GENERADOS:")
    print(f"   üìÇ Directorio: {RESULTS_DIRECTORY}/")
    for model_name in experiment_results.keys():
        print(f"   üìÑ {model_name}.txt")
    
    print("=" * 80)
````
